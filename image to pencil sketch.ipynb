```python
import numpy as np
import cv2
```

```python
from matplotlib import pyplot as plt
```

Taking image input

```python
img = cv2.imread('Shawn-Mendes.jpg')
```

Converting BGR to RGB

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

Converting img to gray scale

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
plt.show()
```

Inverting gray scale img

```python
inverted= ~gray_img
```

```python
plt.imshow(inverted, cmap='gray')
plt.show()
```

Applying Gaussian Blur to remove noise

```python
blurred_img=cv2.GaussianBlur(inverted,(21,21),0)
```

```python
plt.imshow(blurred_img, cmap='gray')
plt.show()
```

Converted img to Pencil Sketch

```python
dodged=cv2.divide(gray_img,255 - blurred_img,scale=256)
```

```python
plt.imshow(dodged, cmap='gray')
plt.show()
```
